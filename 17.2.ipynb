* Use sqlite3 to connect to books database
* Create a cursor object
* use the method execute to select all data in the titles table
* use description and fetchall to display in a table

```python
import sqlite3
```

connect to the database

```python
connection = sqlite3.connect('books.db')
```

create cursor object

```python
cursor = connection.cursor()
```

use execute method to select data in the titles table and look at the description

```python
cursor=cursor.execute('SELECT * FROM titles')
cursor.description
```

The names of the columns need to be extracted and put into a variable. 

```python
columns = [column[0] for column in cursor.description]
```

Now that we have our columns we can:
* use the execute method again to get the data from the titles table
* fetchall to get all the data in the rows
* create a for loop to go through the columns we extracted and print with proper spacing across
* create a for loop to go through the rows data we got when using fetch all and space each piece the same as we did in the colums for loop esuring it also prints across each row until the end
* print our table

```python
cursor = cursor.execute('SELECT * FROM titles')

rows = cursor.fetchall()

for col in columns:
    print(f"{col:<35}", end="")
print()

for row in rows:
    for data in row:
        print(f"{str(data):<35}", end="")
    print()
```
