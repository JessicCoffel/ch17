* Order the output based on the title
  

```python
pd.read_sql("""SELECT  authors.first, authors.last, titles.isbn, titles.title, titles.copyright FROM titles INNER JOIN 
author_ISBN ON titles.isbn = author_ISBN.isbn INNER JOIN authors ON authors.id = author_ISBN.id
WHERE authors.last = 'Deitel' GROUP BY titles.isbn, titles.title, titles.copyright ORDER BY titles.title""", connection)
```
